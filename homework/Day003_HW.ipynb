### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import os
import xmltodict
files=[]
num = 0
for file in os.listdir("./data"):
    if file.endswith("CH.xml"):
        files=files+[os.path.join("./data", file)]

fh = open("./data/64_72hr_CH.xml", "r",encoding="utf-8")
xml = fh.read()
fh.close()
d = dict(xmltodict.parse(xml))
location = d['cwbopendata']['dataset']['locations']['location']

print(len(location))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for loc in location:
    locationName=loc['locationName']
    dataTime=loc['weatherElement'][0]['time'][0]['dataTime']
    value=loc['weatherElement'][0]['time'][0]['elementValue']['value']
    print(locationName+" "+dataTime+" "+value)
    
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

loc=location[0]
locationName=loc['locationName']
len_time=len(loc['weatherElement'][0]['time'])
print(locationName)
for t in range(len_time):
    dataTime=loc['weatherElement'][0]['time'][t]['dataTime']
    value=loc['weatherElement'][0]['time'][t]['elementValue']['value']

    print(dataTime+" "+value)
```
