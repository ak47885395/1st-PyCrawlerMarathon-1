## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
response=r.text
soup=BeautifulSoup(response,"lxml")
p=soup.find_all('a', class_='PostEntry_root_V6g0rd')
print(type(p))
#print(p[0])
for t in p:
    ti=t.find('h3', class_='Title__Text-v196i6-0 gmfDU')
    if(ti):
        print(ti.text)
```

## Grab + PyQuery

```python
from grab import Grab
g = Grab()
resp = g.go('https://www.dcard.tw/f')
resp.body[0:3000]
```

```python
from pyquery import PyQuery as pq
doc = pq(resp.body)
title = doc('a[class=PostEntry_root_V6g0rd]')
print(title)

#print(type(title), title.text())
```

```python
L=len(title)
print(L)
for i in range(L):
    print(title.find('h3').eq(i).text())
    #print('\n')
```
