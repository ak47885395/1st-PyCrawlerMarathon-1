```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
air=table.find_all('td')
count=0
dic={}
for eachair in air:
    
    if('style' in eachair.attrs and eachair['style']=='white-space:nowrap;'):
        print("0",eachair.text)
        name=eachair.text
        dic[name]={}
        count=0
    elif eachair.text=='\xa0' or len(eachair.attrs)!=0:
        continue
    elif count%2==0:
        #print("1",eachair.text)
        date=eachair.text
        dic[name][date]=''
        count+=1
    else:
        count+=1
        #print("2",eachair.text)
        value=eachair.text
        dic[name][date]=value
print(dic)
```
