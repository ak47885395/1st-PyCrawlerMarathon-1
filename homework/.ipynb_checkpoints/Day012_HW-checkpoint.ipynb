### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
r=requests.get(url)
response=r.text
soup = BeautifulSoup(response, "lxml")
#print(soup.prettify())
#find_all  ResultSet
#find      Tag
#for(Tag)  NavigableString
#for(ResultSet) Tag
#find_all()方法返回的结果是包含值的元素列表(可以很多個或一個列表)，而find()函数直接返回结果。
#find_all()方法没有找到目标是返回空列表, find() 方法找不到目标时,返回 None .
p=soup.find('div', class_='part_list_2')
print(len(p))
#print(type(p))
#print(p)
tag="<class 'bs4.element.Tag'>"
i=0
for t in p:
    #
    #print(type(t))
    
    i=i+1
    if(str(type(t)) == tag):
        ti=t.find('em')
        print(ti)
        print("\n")
        if(ti):
            print(ti.text)
#soup.find(class_="part_list_2").find_all('h3')
```
