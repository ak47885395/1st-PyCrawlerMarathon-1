```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
print(soup.find(class_="part_list_2").find_all('h3'))
```

### ① 取出今天所有的發文

```python
#class名稱不完整也可以唷
#https://blog.gtwang.org/programming/python-beautiful-soup-module-scrape-web-pages-tutorial/2/
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find(class_="tag").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
#https://www.programiz.com/python-programming/dictionary
squares = {}
L=[]
dic = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    #print(type(d.find(class_="tag").text))
    cls=d.find(class_="tag").text
    title=d.find(class_="date").text+d.find_all('a')[-1].text
    if( cls in squares):
        squares[cls]=squares[cls]+1
        dic[cls].append(title)
    else:
        squares[cls]=0
        dic[cls]=[title]
    #print(d.find(class_="date").text,d.find(class_="tag").text, d.find_all('a')[-1].text)
print(squares)
print(dic)
```

### ③ 哪一個類別的文章最多

```python
#print(len(dic))
maxnum=0
for L in dic:
    print(len(dic[L]))
    if(len(dic[L])>maxnum):
        maxnum=len(dic[L])
        maxname = L
print('{} {}'.format(maxname,maxnum))
```
