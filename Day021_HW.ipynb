## 根據範例，取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
timeArray = time.localtime(time.time())
otherStyleTime = time.strftime("%Y %m %d %H:%M:%S", timeArray)
print(otherStyleTime,timeArray.tm_mday)

#print(soup.find(class_="part_list_2").find_all('h3'))
while(1):
    
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    last=soup.find(class_="part_list_2").find_all('h3')[-1]
    struct_time = time.strptime(last.find(class_="date").text, "%Y/%m/%d %H:%M")
    #print(last)
    if(timeArray.tm_mday !=struct_time.tm_mday):
        break
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

#for d in soup.find(class_="part_list_2").find_all('h3'):
#    print(d.find(class_="date").text)
#    struct_time = time.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
#    print(struct_time)
for d in soup.find(class_="part_list_2").find_all('h3'):
    
    struct_time = time.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    #print(last)
    if(timeArray.tm_mday !=struct_time.tm_mday):
        break
    print(d.find(class_="date").text,d.find(class_="tag").text, d.find_all('a')[-1].text)
```

```python


day=3*24*60*60
three = time.localtime(time.time()-day)
otherStyleTime = time.strftime("%Y %m %d %H:%M:%S", three)
print(otherStyleTime,three.tm_mday)
selectSite = Select(browser.find_element_by_id("selM"))
selectSite.select_by_value(str(three.tm_mon))
selectYear = Select(browser.find_element_by_id("selD"))
selectYear.select_by_value(str(three.tm_mday))


browser.find_element_by_id("button").click()
```
